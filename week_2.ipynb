```python
#12. Write a function to find min and max values of any column
import pandas as pd

def find_min_max(df, column_name):
    """
    This function returns the minimum and maximum values of a given column in a DataFrame.

    Parameters:
    - df: The DataFrame containing the data.
    - column_name: The column name for which to find the min and max values.

    Returns:
    - min_value: The minimum value in the column.
    - max_value: The maximum value in the column.
    """
    if column_name in df.columns:  # Check if the column exists in the DataFrame
        # Convert the column to numeric, setting non-numeric values to NaN
        df[column_name] = pd.to_numeric(df[column_name], errors='coerce')

        # Find the min and max values of the column
        min_value = df[column_name].min()
        max_value = df[column_name].max()

        # Check if there are NaN values in the column
        if pd.isna(min_value) or pd.isna(max_value):
            return None, f"Error: Column '{column_name}' contains invalid or missing values."
        return min_value, max_value
    else:
        return None, f"Error: Column '{column_name}' not found in the DataFrame."

# Example usage:
df = pd.read_csv("car_sales.csv")  # Replace with your actual file path

# Call the function with a column name
min_val, max_val = find_min_max(df, 'Price in thousands')

if min_val is None or max_val is None:
    print(max_val)  # This will print the error message
else:
    print(f"Minimum Value: {min_val}")
    print(f"Maximum Value: {max_val}")
```
