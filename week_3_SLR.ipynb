```python
#9. Perform Linear Regression (on train data) without using any library and print the coefficients values
def linear_regression(X, y):
    
    n = len(X)
    sum_x = sum(X)
    sum_y = sum(y)
    sum_xy = sum(X[i] * y[i] for i in range(n))
    sum_xx = sum(X[i] ** 2 for i in range(n))
    
    # Calculating slope (m) and intercept (b)
    m = (n * sum_xy - sum_x * sum_y) / (n * sum_xx - sum_x ** 2)
    b = (sum_y - m * sum_x) / n
    
    return m, b

# Example dataset


# Perform Linear Regression
m, b = linear_regression(X_train, y_train)

# Print the coefficients
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
file_path = "TvMarketing.csv"
df = pd.read_csv(file_path)

# 1. Display first 5 rows
print("1.First 5 rows:")
print(df.head())

# 2. Check column data types
print("\n2.Column Data Types:")
print(df.dtypes)

# 3. Check shape of DataFrame
print("\n3.Shape of DataFrame:", df.shape)

# 4. Summary statistics
print("\n4.Summary Statistics:")
print(df.describe())

# 5. Scatter plot
plt.scatter(df['TV'], df['Sales'], color='blue')
plt.xlabel("TV Marketing Budget")
plt.ylabel("Sales")
plt.title("4.TV Marketing Budget vs Sales")
plt.show()

# 6. Prepare X (feature) and y (response)
X = df[['TV']]
y = df['Sales']

# 7. Split Data into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 8. Print the shape of train and test data
print("\n8.Train Data Shape:", X_train.shape, y_train.shape)
print("Test Data Shape:", X_test.shape, y_test.shape)

# 9. Perform Linear Regression without library
n = len(X_train)
sum_x = X_train.sum()[0]
sum_y = y_train.sum()
sum_xy = (X_train['TV'] * y_train).sum()
sum_xx = (X_train['TV'] ** 2).sum()

# Calculate coefficients
m = (n * sum_xy - sum_x * sum_y) / (n * sum_xx - sum_x ** 2)
b = (sum_y - m * sum_x) / n
print("\n9.Manual Linear Regression Coefficients:")
print(f"Slope (m): {m}")
print(f"Intercept (b): {b}")

# 10. Make predictions manually
y_pred_manual = m * X_test['TV'] + b

# 11. Perform Linear Regression using sklearn
model = LinearRegression()
model.fit(X_train, y_train)

# 12. Print intercept and coefficients
print("\n12.Sklearn Linear Regression Coefficients:")
print(f"Intercept: {model.intercept_}")
print(f"Coefficient: {model.coef_[0]}")

# 13. Make predictions using sklearn
y_pred_sklearn = model.predict(X_test)

# 14. Plot train data and manual regression model
plt.scatter(X_train, y_train, color='blue', label='Train Data')
plt.plot(X_train, m * X_train + b, color='red', label='Manual Regression')
plt.xlabel("TV Marketing Budget")
plt.ylabel("Sales")
plt.title("14.Train Data & Manual Regression Model")
plt.legend()
plt.show()

# 15. Plot test data and sklearn model
plt.scatter(X_test, y_test, color='green', label='Test Data')
plt.plot(X_test, model.predict(X_test), color='red', label='Sklearn Regression')
plt.xlabel("TV Marketing Budget")
plt.ylabel("Sales")
plt.title("15.Test Data & Sklearn Regression Model")
plt.legend()
plt.show()

# 16. Print Mean Squared Error and R Squared Value
mse = mean_squared_error(y_test, y_pred_sklearn)
r2 = r2_score(y_test, y_pred_sklearn)
print("\nMean Squared Error:", mse)
print("R Squared Value:", r2)

# 17. Calculate R Squared and Adjusted R Squared without library
ss_total = ((y_test - y_test.mean()) ** 2).sum()
ss_residual = ((y_test - y_pred_manual) ** 2).sum()
r2_manual = 1 - (ss_residual / ss_total)
adj_r2_manual = 1 - ((1 - r2_manual) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1))

print("\nManual R Squared:", r2_manual)
print("Manual Adjusted R Squared:", adj_r2_manual)
```
