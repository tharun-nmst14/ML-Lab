```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# Lets take an small example
x_axis = np.array([1,2,3,18,19,20,0,0,0,0,0,0,1,2,3,18,19,20])
y_axis = np.array([1,2,3,18,19,20,1,2,3,18,19,20,0,0,0,0,0,0])

data = pd.DataFrame({'x':x_axis, 'y':y_axis})

plt.plot()
plt.title('Dataset')
plt.scatter(x_axis,y_axis)
plt.show()
     
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=4,max_iter=100).fit(data)
data['clusters'] = km.labels_
print('Clusters assigned to each datapoints :')
print(data['clusters'])
print('\nSum of Square Error/ WCSS is :',km.inertia_)
```
